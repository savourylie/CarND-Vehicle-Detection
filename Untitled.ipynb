```python
dfCar = dfMain[(dfMain['label'] == 'car') | (dfMain['label'] == 'truck')]
dfNotCar = dfMain[(dfMain['label'] != 'car') & (dfMain['label'] != 'truck')]
```

```python
def get_image(row):
    filename = './data/object-dataset/' + row[0]
    xmin = row[1]
    ymin = row[2]
    xmax = row[3]
    ymax = row[4]
    
    img = plt.imread(filename)
    img = img[ymin:ymax, xmin:xmax]
    
    return img
```

```python
plt.imshow(get_image(dfMain.ix[6, :]))
plt.show()
```

```python
for i, row in dfCar.iterrows():
    cv2.imwrite('./data/udacity/vehicles/car' + str(i) + '.png', get_image(row))

# for i, row in dfNotCar.iterrows():
#     notcar_image_data.append(get_image(row))
```

```python
for i, row in dfNotCar.iterrows():
    cv2.imwrite('./data/udacity/non-vehicles/notcar' + str(i) + '.png', get_image(row))
```

```python
def draw_boxes(img, bboxes, color=(0, 0, 255), thick=6):
    # Make a copy of the image
    imcopy = np.copy(img)
    # Iterate through the bounding boxes
    for bbox in bboxes:
        # Draw a rectangle given bbox coordinates
        cv2.rectangle(imcopy, bbox[0], bbox[1], color, thick)
    # Return the image copy with boxes drawn
    return imcopy
```

```python



row = 3

img = plt.imread('./data/object-dataset/' + dfMain.ix[row, 0])
plt.figure(figsize=(16,9))
plt.imshow(img)

xmin = int(dfMain.ix[row, 1])
ymin = int(dfMain.ix[row, 2])
xmax = int(dfMain.ix[row, 3])
ymax = int(dfMain.ix[row, 4])
img_box = draw_boxes(img, [((xmin, ymin), (xmax, ymax))])
plt.imshow(img_box)
```
